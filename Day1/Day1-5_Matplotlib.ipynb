## ④基本操作

```python
import seaborn as sns

# テーマをセット（デフォルト）
sns.set_theme()

# seaborn内にデフォルトで入っているデータセットを呼び出す
tips = sns.load_dataset("tips")

tips
```

```python
# relplot: 散布図
# data: データセット
# x: X軸に使うカラム名
# y: Y軸に使うカラム名
# col: 比較に使うカラム名
# hue: 色分けに使うカラム名
# style: スタイル（●☓）分けに使うカラム名
# size: 大きさに使うカラム名
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

## ⑤すべての散布図を作成する
- pairplotですべての数値カラムをかけ合わせた散布図が作成されます

```python
# irisのデータセットもseabornに入っています
iris = sns.load_dataset("iris")
iris
```

```python
sns.pairplot(data=iris, hue="species")
```

## ⑥catplot
- 質的変数と量的変数の関係を描画するのに適したインターフェイス

### 主な引数
- data: 参照するデータ
- x,y: プロットするカラム名
- col: 比較に使うカラム名
- hue: 色分けに使うカラム名
- kind: グラフの種類

```python
tips
```

```python
sns.catplot(data=tips, x="day", y="tip", col="time", hue="smoker", kind="swarm")
```

```python
sns.catplot(data=tips, x="day", y="tip", col="time", kind="box")
```
